# Layer 13: Natural Language Querying (Conceptual LLM)

To democratize data access for non-technical users, we conceptualize an LLM wrapper. A user can ask a natural language question, and the system dynamically generates PySpark SQL!

```
def text_to_sql(question):
    # Conceptual LLM translation logic
    print(f"👤 User Question: '{question}'")
    print("🧠 LLM translating to PySpark SQL...")
    
    if "top selling items" in question.lower():
        sql_query = """
        SELECT product_id, COUNT(*) as total_sales 
        FROM user_sessions 
        WHERE label = 1 
        GROUP BY product_id 
        ORDER BY total_sales DESC 
        LIMIT 5
        """
        return sql_query
    return "SELECT * FROM user_sessions LIMIT 5"

# Simulate an execution
generated_query = text_to_sql("Show me the top selling items")
print(f"\n⚙️ Generated Query:\n{generated_query}")

print("\n📊 Executing Query on Delta Lake...")
# spark.sql(generated_query).show()
```
