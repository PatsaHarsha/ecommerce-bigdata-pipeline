# Layer 6a: Machine Learning Training

Here we train a Distributed Random Forest algorithm on our Delta Lake features to predict purchase intent.

```
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.classification import RandomForestClassifier

# Load features
df = spark.read.format("delta").load("s3a://ecommerce-lakehouse/features_prepared/")

# Vectorize features
feature_cols = ["view_count", "cart_count", "total_session_value", "session_duration_seconds"]
assembler = VectorAssembler(inputCols=feature_cols, outputCol="features")
assembled_df = assembler.transform(df)

train_df, test_df = assembled_df.randomSplit([0.8, 0.2], seed=42)

# Train Random Forest
rf = RandomForestClassifier(labelCol="label", featuresCol="features", numTrees=20, maxDepth=5)
model = rf.fit(train_df)
print("Model trained successfully!")
```
