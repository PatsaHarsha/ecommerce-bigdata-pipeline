# Layer 6b: Model Evaluation & MLOps

We track our metrics and save our models to the MLflow Model Registry. Our Random Forest model achieved an impressive `0.8938` ROC-AUC score!

```
import mlflow
from pyspark.ml.evaluation import BinaryClassificationEvaluator

mlflow.set_tracking_uri("http://mlflow:5000")
mlflow.set_experiment("ecommerce_purchase_prediction")

predictions = model.transform(test_df)
evaluator = BinaryClassificationEvaluator(labelCol="label", rawPredictionCol="rawPrediction", metricName="areaUnderROC")
roc_auc = evaluator.evaluate(predictions)

with mlflow.start_run():
    mlflow.log_metric("roc_auc", roc_auc)
    mlflow.log_param("numTrees", 20)
    mlflow.log_param("maxDepth", 5)
    
    # Log model & register it
    mlflow.spark.log_model(model, "model", registered_model_name="PurchasePredictionModel")
    
print(f"Final Model ROC-AUC Score: {roc_auc:.4f}")
```
