# Layer 11: System & Model Monitoring (Grafana + Prometheus)

Our infrastructure is heavily monitored. We use Prometheus to scrape system metrics (like Spark Executor CPU or Kafka lag) and Grafana to visualize them in beautiful dashboards.

```
import requests

try:
    # Query our local Prometheus server running in Docker
    response = requests.get("http://localhost:9090/api/v1/query?query=up")
    metrics = response.json()
    
    print("Prometheus is actively scraping the following targets:")
    for result in metrics["data"]["result"]:
        print(f"✅ Target: {result['metric']['job']} | Instance: {result['metric']['instance']}")
except Exception as e:
    print("Ensure Prometheus container is running on port 9090.")
```
