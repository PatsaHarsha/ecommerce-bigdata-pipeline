# Layer 9: Real-Time Streaming Dashboard

By leveraging Delta Lake's time-travel and streaming read capabilities, we can derive live metrics as our Kafka stream continuously ingests new records.

```
# Streaming query for live purchases
live_purchases = spark.readStream.format("delta").load("s3a://ecommerce-lakehouse/events_raw/") \
    .filter(col("event_type") == "purchase") \
    .groupBy("product_id") \
    .count() \
    .orderBy(col("count").desc())

# In a real dashboard, this writes to a memory sink or a fast key-value store.
# query = live_purchases.writeStream.outputMode("complete").format("memory").queryName("live_purchases").start()
```
