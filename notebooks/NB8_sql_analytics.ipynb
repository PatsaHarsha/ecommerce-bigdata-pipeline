# Layer 8: PySpark SQL Analytics

Beyond Machine Learning, we can extract immense business value from our Delta Tables using PySpark SQL. Below are queries to calculate average session value and overall cart abandonment rates.

```
# Register the Delta Table as a temporary SQL view
df.createOrReplaceTempView("user_sessions")

# 1. Average Session Value
avg_val = spark.sql("SELECT AVG(total_session_value) as avg_session_value FROM user_sessions WHERE total_session_value > 0")
avg_val.show()

# 2. Cart Abandonment Rate
# Sessions with at least 1 cart but 0 purchases
abandonment = spark.sql("""
SELECT 
    SUM(CASE WHEN cart_count > 0 AND label = 0 THEN 1 ELSE 0 END) / 
    SUM(CASE WHEN cart_count > 0 THEN 1 ELSE 0 END) * 100 as cart_abandonment_percentage
FROM user_sessions
""")
abandonment.show()
```
